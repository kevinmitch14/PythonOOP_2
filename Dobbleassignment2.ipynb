```python
from random import choice
import sys

# nIm - 1 must be prime
# Cards must have 3, 4, 6 or 8 images

nIm = 8
n = nIm - 1
r = range(n)
rp1 = range(n+1)
c = 0
deck = dict()
contents = set()


# First card
c += 1
for i in rp1:
    contents.add(i + 1)
deck[c] = contents

# Following cards
for j in r:
    c = c+1
    contents = set()
    contents.add(1)
    for k in r:
        contents.add(n+2 + n*j +k)
        deck[c] = contents

# n x n following cards
for i in r:     
    for j in r:
        c = c+1
        contents = set()
        contents.add(i+2)
        for k in r:
            contents.add((n+1 +n*k + (i*k+j) % n)+1)
            deck[c] = contents


import emoji
imageDict = dict()
fin = open('emoji_names.txt',"r")
lines = fin.readlines()
for i, el in enumerate(lines):
    imageDict[i+1] = emoji.emojize(el.strip())

```
